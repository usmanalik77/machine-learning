```python
dfs=df[['V2','V3','V4','V9','V10','V11','V12','V14','V17','V19','Class']]
dfs.shape 
dffs=dfs.loc[dfs['Class']==1]
dfns=dfs.loc[dfs['Class']==0]
#f_n_frac=len(dff)/len(dfn)
```

```python
#sns.boxplot(x='V2', data=dfs)
#Define statistical quartiles

q1ns=dfns.quantile(0.25)
q3ns=dfns.quantile(0.75)
iqrns=q3ns-q1ns

q1fs=dffs.quantile(0.25)
q3fs=dffs.quantile(0.75)
iqrfs=q3fs-q1fs

iqrfs.head()
iqrfs['V2']
```

```python
#Drop outlier data rows
cols=list(dfs.columns)
cols.pop()
cols
for i in cols: 
    
    dfns=dfns.drop(dfns[dfns[i] > q3ns[i] + 1.5*iqrns[i]].index)
    dffs=dffs.drop(dffs[dffs[i] > q3fs[i] + 1.5*iqrfs[i]].index)
    
    dfns=dfns.drop(dfns[dfns[i] < q1ns[i] - 1.5*iqrns[i]].index)
    dffs=dffs.drop(dffs[dffs[i] < q1fs[i] - 1.5*iqrfs[i]].index)
    
    #dfns[i][dfns[i]> q3ns[i] + 1.5*iqrns[i] ]= np.nan  #dfns[i].mean()
    #dffs[i][dffs[i]> q3fs[i] + 1.5*iqrfs[i] ]= np.nan
    #dfns[i][dfns[i]< q1ns[i] - 1.5*iqrns[i] ]= np.nan
    #dffs[i][dffs[i]< q1fs[i] - 1.5*iqrfs[i] ]= np.nan

```

```python
#combine as trained dataset
df_cleaned=dfns.append(dffs, ignore_index=True)
df_cleaned=df_cleaned.dropna(axis=0)
df_cleaned
```

```python
X=df_cleaned.drop(['Class'], axis=1)
y=df_cleaned['Class']
np.bincount(y)
```

```python
#X=dfs.drop(['Class'], axis=1)
#y=dfs['Class']

#dfs
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression

m1=LogisticRegression(random_state=0)
#fit data
m1.fit(X, y)
y_pred=m1.predict(x_test)


#print(classification_report(Y_test, Y_pred1))
```

```python
#from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
#print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test,y_pred))
#pd.crosstab(y_test,y_pred)

```

```python
## plotting confusion matrix with heatmap

LABELS = ['valid', 'Fraud']
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10, 10))
sns.heatmap(conf_matrix, xticklabels=LABELS,
            yticklabels=LABELS, annot=True, fmt="d");
plt.title("Confusion matrix")
plt.ylabel('Actual classification')
plt.xlabel('Predicted classification')
plt.show()

```

## Repeat with SMOTE

```python
#import and rename functions to run SMOTE
import six
import sys
import sklearn
sys.modules['sklearn.externals.six'] = six
import sklearn.neighbors._base
sys.modules['sklearn.neighbors.base'] = sklearn.neighbors._base
```

```python
from imblearn.over_sampling import SMOTE
oversample = SMOTE()
X_res, y_res = oversample.fit_sample(x_train, y_train)

np.bincount(y_res)
```

```python
from sklearn.linear_model import LogisticRegression

m1_smote=LogisticRegression(random_state=0)
#fit data
m1_smote.fit(X_res, y_res)
y_pred_smote=m1_smote.predict(x_test)
```

```python
#from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report
#print(accuracy_score(y_test,y_pred))
print(confusion_matrix(y_test, y_pred_smote))
print(classification_report(y_test,y_pred_smote))
#pd.crosstab(y_test,y_pred)

```

```python
np.bincount(y_pred_smote)
```

```python
## plotting confusion matrix with heatmap

LABELS = ['valid', 'Fraud']
conf_matrix = confusion_matrix(y_test, y_pred_smote)
plt.figure(figsize=(10, 10))
sns.heatmap(conf_matrix, xticklabels=LABELS,
            yticklabels=LABELS, annot=True, fmt="d");
plt.title("Confusion matrix")
plt.ylabel('Actual classification')
plt.xlabel('Predicted classification')
plt.show()

```
